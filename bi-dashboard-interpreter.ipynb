```python
# ==========================================
# Atualizar Pain√©is
# ==========================================

import json
import shutil
import os
from pathlib import Path
import reporter

# ==========================================
# 1. Defina seus links para atualiza√ß√£o aqui
# ==========================================
minhas_urls = [    
    "https://app.fabric.microsoft.com/view?r=eyJrIjoiYTA3ZDQ3ZmQtYzFlMi00YTYyLTgwNjgtYWU4OGU0YzNmYzc1IiwidCI6IjE2ZGVlZGU4LTc5ZTktNGUzMy05OWU2LTlkOGQzOTQyZDc5NiIsImMiOjEwfQ%3D%3D",
    # Adicione outros links aqui se quiser
]

# Normaliza urls
target_urls = set(u.strip() for u in minhas_urls if u.strip())

# ==========================================
# 2. Checa se j√° existem no "runs/" e deleta (inclusive do cache processed_urls.json)
# ==========================================
runs_dir = Path("runs")
processed_json_path = runs_dir / "processed_urls.json"
cleaned_count = 0
# A) Limpa do arquivo central de mem√≥ria (processed_urls.json)
if processed_json_path.exists():
    try:
        with open(processed_json_path, 'r', encoding='utf-8') as f:
            history = json.load(f)
        
        # Identifica chaves a remover
        keys_to_remove = [url for url in history if url in target_urls]
        
        if keys_to_remove:
            for k in keys_to_remove:
                del history[k]
                print(f"üß† Mem√≥ria limpa para: {k}")
            
            # Salva de volta
            with open(processed_json_path, 'w', encoding='utf-8') as f:
                json.dump(history, f, indent=2)
                
    except Exception as e:
        print(f"‚ö†Ô∏è Erro ao atualizar processed_urls.json: {e}")
# B) Limpa as pastas f√≠sicas
if runs_dir.exists():
    found_files = list(runs_dir.glob("**/catalog_*.json"))
    
    for json_file in found_files:
        try:
            if not json_file.exists(): continue
                
            with open(json_file, 'r', encoding='utf-8') as f:
                data = json.load(f)
            
            saved_url = data.get('url', '').strip()
            
            if saved_url in target_urls:
                folder_to_delete = json_file.parent
                print(f"‚ôªÔ∏è  Encontrado pasta antiga: {folder_to_delete}")
                
                try:
                    shutil.rmtree(folder_to_delete)
                    cleaned_count += 1
                except Exception as e:
                    print(f"    ‚ùå Erro ao deletar pasta: {e}")
                    
        except Exception as e:
            print(f"‚ö†Ô∏è Erro ao processar arquivo {json_file}: {e}")
print(f"\nüßπ Limpeza conclu√≠da: {cleaned_count} pastas removidas.")

# ==========================================
# 3. Reseta a interface gr√°fica
# ==========================================

# Gera o relat√≥rio com os dados atuais da pasta 'runs'. Esse m√©todo j√° √© chamado nos pipelines principais. 
# Se s√≥ quiser remover um painel n√£o precisa rdar o main.py ou batch_main.py.
report_path = reporter.generate_report()
print(f"Relat√≥rio gerado em: {report_path}")

# ==========================================
# 4. Salva urls.json
# ==========================================
with open("urls.json", "w", encoding="utf-8") as f:
    json.dump(list(target_urls), f, indent=2)

print(f"üìù Arquivo 'urls.json' atualizado com {len(target_urls)} URLs.")
print("üöÄ Tudo pronto! Pode rodar o '!python batch_main.py' agora.")

# ==========================================
# 5. Executa o script (eu acho melhor pelo terminal)
# ==========================================
# !python batch_main.py
```
