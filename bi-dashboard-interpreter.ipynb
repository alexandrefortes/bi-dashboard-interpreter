```python
# ==========================================
# Atualizar Pain√©is (Smart Update) - FIXED
# ==========================================

import json
import shutil
import os
from pathlib import Path

# ==========================================
# 1. Defina seus links para atualiza√ß√£o aqui
# ==========================================
minhas_urls = [    
    "https://app.fabric.microsoft.com/view?r=eyJrIjoiYTA3ZDQ3ZmQtYzFlMi00YTYyLTgwNjgtYWU4OGU0YzNmYzc1IiwidCI6IjE2ZGVlZGU4LTc5ZTktNGUzMy05OWU2LTlkOGQzOTQyZDc5NiIsImMiOjEwfQ%3D%3D",
    # Adicione outros links aqui se quiser
]

# Normaliza urls
target_urls = set(u.strip() for u in minhas_urls if u.strip())

# ==========================================
# 2. Checa se j√° existem no "runs/" e deleta
# ==========================================
runs_dir = Path("runs")
cleaned_count = 0

if runs_dir.exists():
    # 1. COLETA (Leitura apenas)
    # Convertemos para list() para for√ßar a leitura de tudo ANTES de mexer em qualquer coisa
    found_files = list(runs_dir.glob("**/catalog_*.json"))
    
    # 2. A√á√ÉO (Agora podemos deletar sem quebrar o loop)
    for json_file in found_files:
        try:
            # Verifica se o arquivo ainda existe (pois podemos ter deletado a pasta pai dele em um passo anterior)
            if not json_file.exists():
                continue
                
            with open(json_file, 'r', encoding='utf-8') as f:
                data = json.load(f)
            
            saved_url = data.get('url', '').strip()
            
            if saved_url in target_urls:
                folder_to_delete = json_file.parent
                print(f"‚ôªÔ∏è  Encontrado processamento antigo para: {saved_url}")
                print(f"    Deletando pasta: {folder_to_delete}")
                
                try:
                    shutil.rmtree(folder_to_delete)
                    cleaned_count += 1
                except Exception as e:
                    print(f"    ‚ùå Erro ao deletar: {e}")
                    
        except Exception as e:
            print(f"‚ö†Ô∏è Erro ao processar arquivo {json_file}: {e}")

print(f"\nüßπ Limpeza conclu√≠da: {cleaned_count} execu√ß√µes antigas removidas.")

# ==========================================
# 3. Reseta a interface gr√°fica
# ==========================================
report_dir = Path("bi_catalog_report")
if report_dir.exists():
    try:
        shutil.rmtree(report_dir)
        print("‚úÖ Pasta 'bi_catalog_report' removida para regera√ß√£o limpa.")
    except Exception as e:
        print(f"‚ùå Erro ao limpar report dir: {e}")

# ==========================================
# 4. Salva urls.json
# ==========================================
with open("urls.json", "w", encoding="utf-8") as f:
    json.dump(list(target_urls), f, indent=2)

print(f"üìù Arquivo 'urls.json' atualizado com {len(target_urls)} URLs.")
print("üöÄ Tudo pronto! Pode rodar o '!python batch_main.py' agora.")

# ==========================================
# 5. Executa o script (eu acho melhor pelo terminal)
# ==========================================
# !python batch_main.py
```
