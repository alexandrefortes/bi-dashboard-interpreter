```python
# ==========================================
# Resetar tudo que foi processado (CUIDADO!)
# ==========================================

import shutil
import os

# Pastas para deletar
folders_to_reset = ['runs']

for folder in folders_to_reset:
    if os.path.exists(folder):
        try:
            shutil.rmtree(folder)
            print(f"✅ Pasta deletada: {folder}")
        except Exception as e:
            print(f"❌ Erro ao deletar {folder}: {e}")
    else:
        print(f"⚠️ Pasta não encontrada (já deletada ou ainda não criada): {folder}")

print("Reset concluído! Pronto para nova execução.")
```
